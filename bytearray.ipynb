# What is bytearray in Python?

bytearray is a **mutable sequence of bytes.**

That means:

- Stores binary data (0 to 255)
- Similar to bytes
- BUT can be modified after creation

# Why bytearray Exists

Normal strings store text.

But computers internally work with:

- binary files
- images
- videos
- packets
- encrypted data

bytearray helps manipulate raw binary data efficiently.

# . Difference Between bytes and bytearray
| Feature                  | bytes | bytearray       |
| ------------------------ | ----- | --------------- |
| Mutable                  | ❌ No  | ✅ Yes           |
| Performance              | Fast  | Slightly slower |
| Can modify data          | ❌     | ✅               |
| Used in hashing/security | ✅     | Sometimes       |
| Memory efficient editing | ❌     | ✅               |


# Creating bytearray

```python
# Method 1 — Empty bytearray
data = bytearray()

print(data)
```

```python
# Method 2 — From list of integers
# Each number must be between 0 and 255

data = bytearray([65, 66, 67])

print(data)
```

```python
# Method 3 — From string
# Convert string to bytes using encoding

data = bytearray("python", "utf-8")

print(data)
```

```python
# Method 4 — Fixed size
# Creates 5 zero bytes

data = bytearray(5)

print(data)
```

# . Important Rule

Every byte value must be:

0 <= value <= 255

Wrong:

bytearray([300])

Error:

ValueError

```python
# Accessing Elements
data = bytearray(b"python")

print(data[0])  # ASCII value
print(chr(data[0]))  # Convert to character
```

```python
# . Modifying Elements

# This is the BIG advantage over bytes.

data = bytearray(b"python")

# Change p -> j
data[0] = ord('j')

print(data)
```

```python
# Slicing
data = bytearray(b"fullstack")

print(data[0:4])
```

# Common Operations

```python
# Append
data = bytearray(b"abc")

# Add one byte
data.append(100)

print(data)
```

```python
# Extend
data = bytearray(b"abc")

data.extend(b"def")

print(data)
```

```python
# Insert
data = bytearray(b"ac")

# Insert ASCII of b
data.insert(1, ord('b'))

print(data)
```

```python
# Remove
data = bytearray(b"abc")

data.remove(ord('b'))

print(data)
```

```python
# Pop
data = bytearray(b"abc")

last = data.pop()

print(last)
print(data)
```

```python
# Reverse
data = bytearray(b"python")

data.reverse()

print(data)
```

```python
# Clear
data = bytearray(b"python")

data.clear()

print(data)
```

# Encoding and Decoding

Very important for backend/API jobs.

```python
# tring → bytearray
text = "Hello"

data = bytearray(text, "utf-8")

print(data)
```

```python
# bytearray → string
data = bytearray(b"Hello")

text = data.decode("utf-8")

print(text)
```

# 13. Memory Efficiency

bytearray is memory efficient because:

**mutable**
avoids creating new objects repeatedly

Bad:

data = b""

data += b"a"

data += b"b"

Creates many new objects.

Better:

data = bytearray()

data.extend(b"a")

data.extend(b"b")

# Real Interview Coding Problems
**Problem 1 — Convert String to Uppercase Using bytearray**

```python
text = "python"

# Convert string to mutable binary
data = bytearray(text, "utf-8")

for i in range(len(data)):

    # If lowercase letter
    if 97 <= data[i] <= 122:

        # Convert to uppercase
        data[i] -= 32

print(data.decode())
```

# Problem 2 — Reverse Binary Data

```python
data = bytearray(b"backend")

data.reverse()

print(data)
```

# Problem 3 — Count Vowels in Binary Text

```python
text = bytearray(b"developer")

vowels = b"aeiou"

count = 0

for ch in text:

    if ch in vowels:
        count += 1

print(count)
```

# Real Company-Level Practice Project
Mini Binary File Processor

```python
# Simulate binary file

file_data = bytearray(b"python backend developer")

# Convert spaces to underscores

for i in range(len(file_data)):

    # ASCII for space = 32
    if file_data[i] == 32:

        # Replace with underscore
        file_data[i] = ord('_')

print(file_data.decode())
```
