# What is bytes in Python?

bytes is a data type used to **store binary data.**

Computers understand only binary:

- 01010101

Python represents binary data using bytes.

## 2. Why bytes Exists

Humans read:

"hello"

Computers process:

b'hello'

Every character becomes a number internally.

Example:
| Character | ASCII Value |
| --------- | ----------- |
| h         | 104         |
| e         | 101         |
| l         | 108         |
| o         | 111     
 
So:

b'hello'

Actually becomes:

[104, 101, 108, 108, 111]   |


# Creating Bytes

```python
# Method 2: Using bytes() Constructor

data = bytes([65,66,67])
print(data)
```

# Encoding and Decoding

This is one of the MOST IMPORTANT interview topics.

```python
# Encoding

# Convert string → bytes

text = "hello"

binary = text.encode("utf-8")

print(binary)
```

```python
# Decoding

# Convert bytes → string

data = b'hello'

text = data.decode("utf-8")

print(text)
```

```python
# UTF-8 Encoding

# Most applications use UTF-8.

text = "Python"

data = text.encode("utf-8")

print(data)
```

# ASCII vs UTF-8
| Encoding | Supports      |
| -------- | ------------- |
| ASCII    | English only  |
| UTF-8    | All languages |

# . bytes vs str
| Feature | str       | bytes       |
| ------- | --------- | ----------- |
| Stores  | Text      | Binary data |
| Mutable | No        | No          |
| Prefix  | none      | b           |
| Example | `"hello"` | `b"hello"`  |


```python
# bytearray

# Mutable version of bytes.

data = bytearray(b"hello")

data[0] = 72

print(data)
```

```python
# Mini File Encryption
message = "backend developer"

# Convert string to bytes
data = message.encode()

encrypted = []

# XOR encryption
for byte in data:
    encrypted.append(byte ^ 5)

print(encrypted)
```
