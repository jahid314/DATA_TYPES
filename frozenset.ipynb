# What is frozenset in Python?
A frozenset is an **immutable version of a Python set.**

- set → mutable (can change)
- frozenset → immutable (cannot change)

**That means:**

- You cannot add items
- You cannot remove items
- You cannot update items

**Because it is immutable:**

- It is hashable
- It can be used as:
- dictionary keys
- elements inside another set
- cache keys
- security-safe collections

Use frozenset

In real-world applications, frozenset is used for:

✅ Permission systems ✅ Caching systems ✅ API response validation ✅ Configuration constants ✅ Unique immutable collections ✅ Database query optimization ✅ Security-safe role management ✅ Fast membership testing ✅ Machine learning feature locking

# Operations in frozenset

Although immutable, it supports many set operations.

```python
# 1. Union operation
# combine value
backend = frozenset(["Python", "Django"])
frontend = frozenset(["React", "JavaScript"])

# Combine both sets
all_skills = backend.union(frontend)

print(all_skills)
```

```python
# 2. Intersection Operation
# find common values
team1 = frozenset(["Python", "SQL", "Docker"])
team2 = frozenset(["Python", "AWS", "Docker"])

# Find common technologies
common = team1.intersection(team2)

print(common)
```

```python
# 3.Difference operation
#find unique items
required = frozenset(["Python", "Django", "SQL"])
known = frozenset(["Python", "SQL"])

missing = required.difference(known)

print(missing)
```

```python
# 4. Symmetric Difference

# Items not common in both.

a = frozenset([1, 2, 3])
b = frozenset([3, 4, 5])

result = a.symmetric_difference(b)

print(result)
```

```python
# 5. subset
skills = frozenset(["Python", "SQL"])
required = frozenset(["Python", "SQL", "Django"])

print(skills.issubset(required))
```

```python
# 6.Superset
required = frozenset(["Python", "SQL", "Django"])
skills = frozenset(["Python", "SQL"])

print(required.issuperset(skills))
```

```python
# 7. Disjoint

# Checks whether common elements exist.

backend = frozenset(["Python", "Django"])
mobile = frozenset(["Flutter", "Dart"])

print(backend.isdisjoint(mobile))
```

```python
# Operators with frozenset
a = frozenset([1, 2, 3])
b = frozenset([3, 4, 5])

print(a | b)
print(a & b)
print(a - b)
print(a ^ b)
```

```python
# Iterating Through frozenset
languages = frozenset(["Python", "Java", "Go"])

for lang in languages:
    # Print each item
    print(lang)
    
```

```python
#Membership Testing

#Very important in backend development.
permissions = frozenset(["read", "write", "delete"])

# Fast lookup operation
print("write" in permissions)
```

```python
# Example 1: User Role Permissions
# Immutable permission set
ADMIN_PERMISSIONS = frozenset([
    "create_user",
    "delete_user",
    "update_user",
    "view_reports"
])

# Check permission
user_action = "delete_user"

if user_action in ADMIN_PERMISSIONS:
    print("Access Granted")
else:
    print("Access Denied")
```

```python
# Example 2: API Allowed HTTP Methods

ALLOWED_METHODS = frozenset(["GET", "POST", "PUT", "DELETE"])

request_method = "POST"

if request_method in ALLOWED_METHODS:
    print("Valid Request")
else:
    print("Invalid Request")
```

```python
# Example 3: Cache Key System
# Cache system using immutable key
cache_key = frozenset([
    ("user_id", 101),
    ("page", 2)
])

cache = {}

# Store cached response
cache[cache_key] = "User Data"

print(cache)

# Normal set cannot be dictionary key. frozenset can.
```

```python
# Example 4: Store Unique Tech Stack
company_stack = frozenset([
    "Python",
    "Django",
    "PostgreSQL",
    "Redis"
])

print(company_stack)
```

```python
# Example 5: JWT Role Validation

VALID_ROLES = frozenset([
    "admin",
    "developer",
    "tester"
])

role = "developer"

if role in VALID_ROLES:
    print("Role Verified")
```

# Nested frozenset

A frozenset can be placed inside another set.

```python
skills1 = frozenset(["Python", "Django"])
skills2 = frozenset(["React", "Node"])

teams = {skills1, skills2}

print(teams)
```

# Role-Based Access Control (RBAC)

```python
# immutable role premiision mapping

ROLE_PERMISSIONS = {
    "admin": frozenset([
        "create",
        "update",
        "delete",
        "view"
    ]),

    "developer": frozenset([
        "update",
        "view"
    ]),

    "tester": frozenset([
        "view"
    ])
}

user_role = "developer"
user_action = "delete"

# Get permissions of user role
permissions = ROLE_PERMISSIONS[user_role]

# Check access
if user_action in permissions:
    print("Access Allowed")
else:
    print("Access Denied")
```
