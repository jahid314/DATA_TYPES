# What is a Set in Python?

**A set is:**

- Unordered
- Mutable
- Stores unique values only
- Very fast for searching

```python
numbers = {1, 2, 3, 4}

print(numbers)
```

# 2. Important Characteristics of Sets
| Feature      | Description                |
| ------------ | -------------------------- |
| Unordered    | No indexing                |
| Unique       | No duplicates              |
| Mutable      | Can add/remove items       |
| Fast         | Lookup is very fast        |
| Uses Hashing | Internally uses hash table |

# Set Operations

## A. Union

**Combine all unique elements.**

```python
frontend = {"html","css"}
backend = {"Python","django","css"}
all_skill = frontend.union(backend)
print(all_skill)
```

# B. Intersection

**Common elements.**

```python
python_dev ={"jahid","ali","john"}
django_dev ={"ali","john","mike"}
common = python_dev.intersection(django_dev)
print(common)
```

```python
print(python_dev & django_dev)
```

# C. Difference

**Elements in first set but not second.**

```python
all_users = {"admin","editor","viewer"}
active_users = {"admin","viewer"}
inactive = all_users.difference(active_users)

print(inactive)
```

```python
print(all_users - active_users)
```

# D. Symmetric Difference

**Elements in both but NOT common.**

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.symmetric_difference(b))
```

```python
print(a ^ b)
```

```python
# 13. Subset & Superset
# issubset()

a = {1, 2}
b = {1, 2, 3, 4}

print(a.issubset(b))
```

```python
print(b.issuperset(a))
```

| List              | Set              |
| ----------------- | ---------------- |
| Ordered           | Unordered        |
| Allows duplicates | Unique only      |
| Slower searching  | Faster searching |
| Supports indexing | No indexing      |

# 18. Advanced Internal Concept (Important for Senior Roles)

Python sets internally use:

**Hashing**

When element added:

- Python calculates hash value
- Stores value in hash table
- Lookup becomes extremely fast

```python
# 19. Performance Comparison
# LIST SEARCH
numbers = [1,2,3,4,5,6]

print(6 in numbers)

#List search:

#O(n)
```

```python
# SET SEARCH
numbers = {1,2,3,4,5,6}

print(6 in numbers)

# Set search:

# O(1)
```

```python
#Problem 4: Detect Duplicate Characters
text = "developer"

seen = set()

for char in text:

    # Check if already exists
    if char in seen:
        print("Duplicate:", char)

    # Add character
    seen.add(char)
```
