# What is a List?

**A list** is a built-in Python data structure **used to store multiple items in a single variable.**

```python
my_list = [1, 2, 3, 4]
my_list
```

# Key Features:
- Ordered ✅
- Mutable (changeable) ✅
- Allows duplicate values ✅
- Can store multiple data types ✅

```python
mixed = [1, "hello", 3.5, True]
mixed
```

# Creating Lists

```python
# Empty list
a = []

# With values
b = [10, 20, 30]

# Using list()
c = list((1, 2, 3))

print(f"Empty list:",a)
print(f"with values:",b)
print(f"using List():",c)
```

# Indexing & Accessing Elements


```python
a = [12,34,54,45,67,78]
# a[1]
print(f"Accessing the element using indexing:",a[4])
```

# 4. Slicing


```python

a = [1,2,34,56,76,76,89,9]
print(a[0:5])
print(a[:3])
print(a[::2])
```

# Modifying Lists


```python
a = [1,2,3]

a[0] = 100
print(a)
```

# Adding Elements



```python
a = [1,2]
a.append(3)  # Add at end
a.insert(1,10) # Insert At index

print(a)  
```

# Removing Elements

```python
a = [1,2,3,4]
a.remove(2)  # remove by value

a.pop()  # Remove list

a.pop(1)  # Remove by index

del a[0]  # Delete

print(a)
```

# Looping Through Lists:

```python
a = [1,2,3]

for item in a:
    print(item)
```

```python
# with index

for i in range(len(a)):
    print(i,a[i])
```

# List Comprehension:

```python
# normal Way
squares = []
for i in range(5):
    squares.append(i*i)

print("normal way :",i)

# Pythonic way 
square = [i*i for i in range(5)]

print("Pythonic way :",square)

# With condition:

even = [x for x in range(10) if x % 2 == 0]
print(even)
```

# List Methods

```python
a = [3, 1, 4]

a.append(5)
a.extend([6, 7])
a.sort()
a.reverse()
a.count(3)
a.index(4)
a.copy()
a.clear()
```

# . Built-in Functions

```python
a = [1, 2, 3, 4]

len(a)
max(a)
min(a)
sum(a)
sorted(a)
```

# Nested Lists (2D Lists)

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6]
]

print(matrix[0][1])  # 2
```

# Copying Lists

```python
a = [1, 2, 3]
b = a          # Same reference ❌

c = a.copy()   # New copy ✅
d = a[:]       # Another way ✅
```

# Membership Operators

```python
a = [1, 2, 3]

print(2 in a)      # True
print(5 not in a)  # True
```

# List Packing & Unpacking z) z)

```python
a = [1, 2, 3]

x, y, z = a
print(x, y, z)
```

# Sorting Lists

```python
a = [3, 1, 2]

a.sort()               # Ascending
a.sort(reverse=True)   # Descending

# Custom sort
a = ["apple", "banana", "kiwi"]
a.sort(key=len)
```

# Advanced Concepts (Important for Data Science

## Flatten List

```python
nested = [[1,2],[3,4]]
flat = [item for sublist in nested for item in sublist]
```

# Remove Duplicates

```python
a = [1,2,2,3]
unique = list(set(a))
```

```python
a = [1,2,3,4]
even = list(filter(lambda x: x%2==0, a))
```
