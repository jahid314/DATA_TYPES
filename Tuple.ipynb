# What is Tuple in Python?

A tuple is an **ordered collection of items.**

- Immutable (cannot change after creation)
- Allows duplicate values
- Ordered
- Faster than list
- Uses parentheses ()

```python
#Creating Tuples
# Empty tuple
empty_tuple = ()

print(empty_tuple)
```

```python
#Tuple with values

# Tuple storing different data types
student = ("Jahid", 21, "Python Developer")

print(student)
```

| Usage                     | Example                  |

| Coordinates               | `(x, y)`                 |

# Difference Between List and Tuple

| Feature          | List         | Tuple      |
| ---------------- | ------------ | ---------- |
| Mutable          | ✅ Yes        | ❌ No       |
| Syntax           | `[]`         | `()`       |
| Faster           | ❌            | ✅          |
| Memory efficient | ❌            | ✅          |
| Can modify       | ✅            | ❌          |
| Used for         | Dynamic data | Fixed data |

# Tuple Packing

Packing means storing multiple values into tuple.

```python
# Packing values into tuple
student = ("Jahid", 21, "Python")

print(student)
```

# Tuple Unpacking

VERY IMPORTANT for interviews.

```python
student =("Jahid",21,"Python")

name,age,skill = student

print(name)
print(age)
print(skill)
```

# Real Backend Example of Unpacking

```python
# Function returning multiple values
def get_user():
    
    # Returning tuple automatically
    return "Jahid", "jahid@gmail.com"


# Unpacking returned tuple
name, email = get_user()

print(name)
print(email)
```

# Using * Operator in Unpacking

```python
numbers = (10, 20, 30, 40, 50)

# * collects remaining values
a, b, *others = numbers

print(a)
print(b)
print(others)
```

# Tuple Methods

Tuple has only 2 methods because it is immutable.

| Method  | Purpose           |
| ------- | ----------------- |
| count() | Count occurrences |
| index() | Find position     |


```python
numbers = (10, 20, 10, 30)

# Count how many times 10 appears
print(numbers.count(10))
```

```python
# index()
numbers = (10, 20, 30)

# Find index of 20
print(numbers.index(20))
```

Answer:

- Immutable
- Less memory
- Fixed size

# Q2: Can tuple contain mutable objects?

YES.

```python
data = ([1, 2], [3, 4])

# List inside tuple can change
data[0].append(100)

print(data)
```

# Q3: Is tuple completely immutable?

No.

Tuple itself immutable, but mutable objects inside it can change.

# Advanced Tuple Concepts


```python
# Tuple Packing & Unpacking in Loops

students = [
    ("Jahid", 90),
    ("Rahul", 85),
    ("Aman", 95)
]

# Unpacking inside loop
for name, marks in students:
    
    print(name, marks)
```

```python
#Swapping Variables Using Tuple

# VERY IMPORTANT INTERVIEW QUESTION

a = 10
b = 20

# Tuple unpacking swap
a, b = b, a

print(a)
print(b)
```

```python
# Returning Multiple Values
def calculate(a, b):
    
    # Returning multiple values as tuple
    return a + b, a - b


result = calculate(10, 5)

print(result)
```

# Mini Real-World Project
## Employee Record System

```python
# Employee database records
employees = (
    (101, "Jahid", "Python Developer"),
    (102, "Rahul", "Backend Developer"),
    (103, "Aman", "Frontend Developer")
)

# Loop through employee records
for emp_id, name, role in employees:
    
    # Print formatted employee data
    print(f"ID: {emp_id}")
    print(f"Name: {name}")
    print(f"Role: {role}")
    
    print("---------------")
```
