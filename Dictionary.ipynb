# 2. Features of Dictionary
| Feature           | Description               |
| ----------------- | ------------------------- |
| Mutable           | Can change data           |
| Ordered           | Maintains insertion order |
| Key-Value Pair    | Stores related data       |
| Fast              | Very fast lookup          |

```python
#Accessing Values
student = {
    "name": "Jahid",
    "age": 22
}

# Access value using key
print(student["name"])
```

```python
# Using get() (VERY IMPORTANT)

# Professional developers prefer get().

# Why? Because it prevents errors.
student = {
    "name": "Jahid"
}

# get() avoids KeyError
print(student.get("name"))

# If key does not exist
print(student.get("age"))

```

```python
#Default Value
print(student.get("age", 0))
```

# Difference between [ ] and get( )
| `[]`                            | `get()`            |
| ------------------------------- | ------------------ |
| Raises error if key missing     | Returns None       |
| Faster                          | Safer              |
| Used when key definitely exists | Used for APIs/data |


```python
# Adding New Data
employee = {}

# Add new key-value pair
employee["name"] = "Jahid"
employee["salary"] = 70000

print(employee)
```

```python
#Updating Dictionary
employee = {
    "salary": 50000
}

# Update value
employee["salary"] = 80000

print(employee)
```

```python
# Removing Data
# pop()
student = {
    "name": "Jahid",
    "age": 22
}

# Remove key and return value
removed = student.pop("age")

print(removed)
print(student)
```

```python
# del
student = {
    "name": "Jahid",
    "age": 22
}

# Delete key
del student["age"]

print(student)
```

```python
# clear()
student.clear()

print(student)
```

# Dictionary Methods

```python
# keys()
student = {
    "name": "Jahid",
    "age": 22
}

print(student.keys())
```

```python
#values()
print(student.values())
```

```python
#items()

print(student.items())
```

# Looping Through Dictionary


```python
# Loop Keys
student = {
    "name": "Jahid",
    "age": 22
}

for key in student:
    print(key)
```

```python
# Loop Values
for value in student.values():
    print(value)
```

```python
# Loop Key + Value



for key, value in student.items():
    print(key, value)
```

# Dictionary Comprehension

Advanced and heavily used.

```python
# Basic Example
numbers = {
    x: x*x
    for x in range(5)
}

print(numbers)
```

```python
# Nested Dictionary

# Used heavily in APIs and databases.

students = {
    101: {
        "name": "Jahid",
        "marks": 90
    },

    102: {
        "name": "Rahul",
        "marks": 85
    }
}

print(students[101]["name"])
```

# Dictionary vs JSON
| Dictionary                | JSON               |
| ------------------------- | ------------------ |
| Python object             | Data format        |
| Uses single/double quotes | Uses double quotes |
| Supports Python types     | Text format        |
| Mutable                   | Used for APIs      |



```python
# Convert Dictionary to JSON

# Used in backend development.

import json

user = {
    "name": "Jahid",
    "age": 22
}

# Convert dict to JSON
json_data = json.dumps(user)

print(json_data)
```

```python
# Convert JSON to Dictionary
import json

data = '{"name":"Jahid","age":22}'

# Convert JSON to dict
python_dict = json.loads(data)

print(python_dict)
```

# User Login System

```python
# Database-like dictionary
users = {
    "jahid": "1234",
    "admin": "admin123"
}

username = input("Enter username: ")
password = input("Enter password: ")

# Check login
if username in users:
    
    # Compare password
    if users[username] == password:
        print("Login Successful")
    else:
        print("Wrong Password")

else:
    print("User not found")
```

```python
# Counting Words

# Very common interview problem.
sentence = "python is good and python is powerful"

words = sentence.split()

frequency = {}

for word in words:

    # If word already exists
    if word in frequency:
        frequency[word] += 1

    else:
        frequency[word] = 1

print(frequency)
```

```python
# Using setdefault()

student = {}

# Add only if key not exists
student.setdefault("name", "Jahid")

print(student)
```

```python
# Using fromkeys()
keys = ["name", "age", "salary"]

data = dict.fromkeys(keys, 0)

print(data)
```

# Sorting Dictionary

```python
# Sort by Key
student = {
    "c": 3,
    "a": 1,
    "b": 2
}

sorted_data = dict(sorted(student.items()))

print(sorted_data)
```

```python
# Sort by Value
student = {
    "A": 80,
    "B": 95,
    "C": 70
}

sorted_marks = dict(
    sorted(student.items(), key=lambda x: x[1])
)

print(sorted_marks)
```
