# What is range() in Python?

**range() is a built-in immutable sequence type.**

It generates a sequence of numbers lazily.

That means:

- ✅ It does NOT store all numbers in memory
- ✅ It generates values when needed

# Syntax of range()
Basic Syntax

***range(start, stop, step)***

| Parameter | Meaning                 |
| --------- | ----------------------- |
| start     | Starting value          |
| stop      | Ending value (excluded) |
| step      | Increment/decrement     |


# Types of range() Usage

```python
# Single Argument
numbers = range(5)

print(numbers)
print(list(numbers))
```

```python
# Two Arguments
numbers = range(2, 8)

print(list(numbers))
```

```python
# Three Arguments (Step)
numbers = range(1, 10, 2)

print(list(numbers))
```

```python
# Negative Step

# Used for reverse looping.

numbers = range(10, 0, -1)

print(list(numbers))
```

# Internal Working of range()

range() does NOT create all values immediately.

Example:

r = range(1000000000)

This does NOT consume huge memory.

Why?

Because Python stores only:

(start, stop, step)

This is called:

Lazy Evaluation

Very important interview concept.

```python
# Convert Range to List
numbers = range(5)

print(list(numbers))
```

```python
# Accessing Elements
r = range(10)

print(r[0])
print(r[5])
print(r[-1])
```

```python
# Slicing in Range
r = range(20)

print(r[2:10:2])
print(list(r[2:10:2]))
```

```python
# Checking Membership
r = range(1, 20)

print(10 in r)
print(25 in r)
```

```python
# Length of Range
r = range(1, 100)

print(len(r))
```

# Looping with range()

```python
# Basic Loop
for i in range(5):
    
    # Prints numbers from 0 to 4
    print(i)
```

```python
# Reverse Loop
for i in range(10, 0, -1):
    
    # Reverse counting
    print(i)
```

# enumerate() vs range()

```python
# Bad Way
names = ["A", "B", "C"]

for i in range(len(names)):
    
    print(i, names[i])
```

```python
# Better Way
names = ["A", "B", "C"]

for index, value in enumerate(names):
    
    print(index, value)

# Interview Point

# Use enumerate() when you need:

# index
# value

# Together.
```
